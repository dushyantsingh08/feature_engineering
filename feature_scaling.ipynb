```python
#standardization prefered for ml
#z-score =  x-mue/sigma   mue= 0 , sigma = 1

import seaborn as sns
import numpy as np
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
total_bill = list(df['total_bill'])
```

```python
sns.histplot(total_bill)
```

```python
mean = np.mean(total_bill)
std = np.std(total_bill)
```

```python
mean,std
```

```python
normalized_data =[]
for i in total_bill:
    z_score = (i-mean)/std
    normalized_data.append(z_score)

print(normalized_data)
```

```python
sns.histplot(normalized_data)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler
```

```python
scaler.fit(df[["total_bill",'tip']])
```

```python
scaler.transform(df[["total_bill",'tip']])
```

```python
import pandas as pd
pd.DataFrame(scaler.transform(df[["total_bill",'tip']]),columns=['scaled_bill',"scaled_tip"])
```
