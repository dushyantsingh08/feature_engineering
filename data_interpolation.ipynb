# data interpolation


```python
## linear interpolation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
```

```python
plt.scatter(x,y)
```

```python
x_new  = np.linspace(1,5,10)
y_interp = np.interp(x_new,x,y)
```

```python
plt.scatter(x_new,y_interp)
```

```python
#cubic interpolation
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])
```

```python
from scipy.interpolate import interp1d
func= interp1d(x,y,kind='cubic') # func is kind of function which is created fron first x,y and when we change x -> x_new we an get t_interp cvalue based on that cucbic or xyz func 
```

```python
x_new = np.linspace(1,5,10) # input value change to make more point and get correspoinding y cordinate
y_interp = func(x_new)
print(y_interp)
```

```python
plt.scatter(x_new,y_interp)
```

```python
#polynomial interpolation
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])
```

```python
poly_func = np.polyfit(x,y,2)# 2 is degree of polynomial
```

```python
x_new = np.linspace(1,5,10)
y_interp = np.polyval(poly_func,x_new)
```

```python
plt.scatter(x_new,y_interp)
```
